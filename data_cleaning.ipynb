```python
# Google Colab Code for Cleaning dirty_cafe_sales.csv

import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv('/content/dirty_cafe_sales.csv')

# Display first 5 rows
print("Original Data:")
print(df.head())


# STEP 1: Replace ERROR and UNKNOWN with NaN


df.replace(['ERROR', 'UNKNOWN'], np.nan, inplace=True)


# STEP 2: Clean Quantity Column


df['Quantity'] = pd.to_numeric(df['Quantity'], errors='coerce')

# Fill missing quantity with median
df['Quantity'].fillna(df['Quantity'].median(), inplace=True)


# STEP 3: Clean Price Per Unit Column


df['Price Per Unit'] = pd.to_numeric(df['Price Per Unit'], errors='coerce')

# Fill missing prices with median
df['Price Per Unit'].fillna(df['Price Per Unit'].median(), inplace=True)

# STEP 4: Clean Total Spent Column


df['Total Spent'] = pd.to_numeric(df['Total Spent'], errors='coerce')

# Recalculate missing Total Spent
df['Total Spent'].fillna(
    df['Quantity'] * df['Price Per Unit'],
    inplace=True
)


# STEP 5: Clean Item Column


df['Item'].fillna(df['Item'].mode()[0], inplace=True)


# STEP 6: Clean Payment Method Column


df['Payment Method'].fillna(
    df['Payment Method'].mode()[0],
    inplace=True
)

# STEP 7: Clean Location Column


df['Location'].fillna(
    df['Location'].mode()[0],
    inplace=True
)


# STEP 8: Clean Transaction Date Column


df['Transaction Date'] = pd.to_datetime(
    df['Transaction Date'],
    errors='coerce'
)

# Fill missing dates with forward fill
df['Transaction Date'].fillna(method='ffill', inplace=True)


# STEP 9: Remove Duplicate Rows


df.drop_duplicates(inplace=True)

# STEP 10: Save Cleaned Data


df.to_csv('/content/cleaned_cafe_sales.csv', index=False)

print("\nData Cleaning Completed Successfully!")
print("\nCleaned Data:")
print(df.head())

print("\nCleaned file saved as:")
print("/content/cleaned_cafe_sales.csv")
```
